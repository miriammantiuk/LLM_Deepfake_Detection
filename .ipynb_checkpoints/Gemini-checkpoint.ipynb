```python
import google.generativeai as genai
import os

# Setze deinen API-Key als Umgebungsvariable
# (z.B. im Terminal: export GOOGLE_API_KEY="DEIN_KEY")
try:
    genai.configure(api_key=os.environ["GOOGLE_API_KEY"])
except KeyError:
    print("Fehler: GOOGLE_API_KEY Umgebungsvariable nicht gesetzt.")
    exit()

# WICHTIG: Verwende das "-live" Modell für sofortige Antworten
# (Basierend auf den Tabs in deinem Screenshot)
model_name = 'gemini-2.5-flash' 

try:
    # 1. Initialisiere das Modell
    model = genai.GenerativeModel(model_name)

    # 2. Sende den Prompt
    prompt = "Explain how AI works in a few words"
    response = model.generate_content(prompt)

    # 3. Gib die Antwort aus
    print(response.text)

except Exception as e:
    print(f"Ein Fehler ist aufgetreten:")
    print(e)
    print("\nStelle sicher, dass das Modell '{model_name}' korrekt ist und dein API-Key funktioniert.")
```

```python
import google.generativeai as genai
import os

try:
    # 1. Versuche, den Schlüssel aus der Umgebung zu laden und zu konfigurieren
    genai.configure(api_key=os.environ["GOOGLE_API_KEY"])
    print("API-Konfiguration erfolgreich.")
    
    # 2. Mache einen echten Test-Anruf (z.B. Modelle auflisten)
    print("Teste API-Verbindung durch Auflisten der Modelle...")
    genai.list_models() # Dieser Anruf schlägt fehl, wenn der Key ungültig ist
    
    print("\n✅ TEST ERFOLGREICH! Der API-Schlüssel ist gesetzt UND gültig.")

except KeyError:
    print("\n❌ FEHLER (KeyError): Der Schlüssel wurde nicht in der Umgebung gefunden (siehe Test 1).")
except Exception as e:
    # Dies fängt Authentifizierungsfehler ab
    print(f"\n❌ FEHLER (Exception): Der Schlüssel wurde gefunden, aber der Anruf schlug fehl.")
    print("   Das bedeutet meist, dass der Schlüssel selbst ungültig, abgelaufen oder falsch kopiert ist.")
    print(f"   Fehlerdetails: {e}")
```
