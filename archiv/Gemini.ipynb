```python
import google.generativeai as genai
import os
import time
from pathlib import Path  # <-- Sehr nützlich für die Arbeit mit Ordnern

# --- KONFIGURATION (Plan B) ---
MY_API_KEY = "AIzaSyDGP8zC9ocr31AnZXtZBcR2F_2Yu8h93aA"  # ERSETZE DAS
genai.configure(api_key=MY_API_KEY)
print("Konfiguration erfolgreich.")

# 1. Definiere den Ordner, in dem deine ENTPACKTEN Videos liegen
#    (Passe den Pfad an, wohin du die zip-Datei entpackt hast)
video_ordner = Path(r"C:\Users\miria\OneDrive - IU International University of Applied Sciences\Masterarbeit\videos")

# 2. Wähle das Modell (nur einmal definieren)
model = genai.GenerativeModel("gemini-2.5-pro")

# 3. Schleife durch alle mp4-Dateien in diesem Ordner
for video_pfad in video_ordner.glob("*.mp4"):
    print(f"\n--- Beginne Analyse für: {video_pfad.name} ---")
    
    try:
        # 4. Lade JEDES Video einzeln hoch
        print("Lade Video hoch...")
        video_file = genai.upload_file(path=video_pfad)
        
        # 5. Warte auf die Verarbeitung
        print("Video wird verarbeitet...")
        while video_file.state.name == "PROCESSING":
            time.sleep(10)
            video_file = genai.get_file(video_file.name)
        
        if video_file.state.name == "FAILED":
            print(f"Verarbeitung für {video_pfad.name} fehlgeschlagen.")
            continue  # Mache mit dem nächsten Video weiter
            
        print("Video ist bereit.")
        
        # 6. Analysiere das Video (Passe den Prompt für deine Forschung an)
        prompt = """You are an expert in deepfake detection. Analyze the video using the following indicators:
                    Unnatural body movements, stiff or robotic gestures, missing or inconsistent facial expressions
                    Facial irregularities such as uneven skin tones, poor eye rendering, asymmetry issues, or unnatural textures
                    Inconsistencies in lighting, shadows, colors, or reflections
                    Signs of face swapping, unnatural lip or mouth movements, abnormal eye movements or blinking patterns
                    Visual artifacts: blurring, pixel noise, edge glitches, compression issues, resolution inconsistencies- Frequency anomalies or generator/compression-related patterns
                    Temporal instabilities: warping, frame jumps, inconsistent or changing details between frames
                    Logical coherence and contextual consistency of the visual content
                    Provide the result in the following format:
                    Assessment (Real | Fake)
                    Probability in % (Real/Fake)
                    Justification (max. 5 sentences), based on the indicators above.
                    """
        
        response = model.generate_content([prompt, video_file])
        
        print(f"Antwort für {video_pfad.name}:\n{response.text}\n")
        
        # 7. Aufräumen (Datei vom Google-Server löschen)
        genai.delete_file(video_file.name)
        print("Datei vom Server gelöscht.")

    except Exception as e:
        print(f"Ein unerwarteter Fehler bei {video_pfad.name}: {e}")
        # Beende die Schleife nicht, mache mit dem nächsten Video weiter
        continue

print("\n--- Alle Videos im Ordner analysiert ---")
```

```python
import cv2
import os
from pathlib import Path

def extract_frames_from_first_n_videos(source_folder_path, output_base_folder, num_videos=2, frame_interval=10):
    source_path = Path(source_folder_path)
    
    # 1. Alle Videos finden und sortieren
    all_videos = sorted(list(source_path.glob("*.mp4")))
    
    if not all_videos:
        print(f"❌ Keine .mp4 Videos im Ordner gefunden.")
        return

    # 2. Wähle nur die ersten N Videos aus (hier: 2)
    # Das [:num_videos] ist der Python-Trick (Slicing)
    videos_to_process = all_videos[:num_videos]
    
    print(f"🎬 Gefunden: {len(all_videos)} Videos. Verarbeite die ersten {len(videos_to_process)}...\n")

    # 3. Schleife durch die ausgewählten Videos
    for i, video_path in enumerate(videos_to_process):
        video_name = video_path.name
        video_stem = video_path.stem # Name ohne .mp4
        
        print(f"--- Starte Video {i+1} von {len(videos_to_process)}: {video_name} ---")
        
        # Zielordner für DIESES spezifische Video erstellen
        target_folder = os.path.join(output_base_folder, video_stem)
        if not os.path.exists(target_folder):
            os.makedirs(target_folder)

        # Capture starten
        cap = cv2.VideoCapture(str(video_path))
        
        if not cap.isOpened():
            print(f"❌ Konnte {video_name} nicht öffnen. Überspringe...")
            continue

        count = 0
        saved_count = 0
        
        while True:
            ret, frame = cap.read()
            if not ret:
                break

            # Speichern (mit frame_interval)
            if count % frame_interval == 0:
                filename = os.path.join(target_folder, f"frame_{count:04d}.jpg")
                cv2.imwrite(filename, frame)
                saved_count += 1
            count += 1

        cap.release()
        print(f"✅ Fertig mit Video {i+1}. {saved_count} Bilder gespeichert in: /{video_stem}")

# --- DEINE PFADE ---

input_folder = r"C:\Users\miria\Deepfake_Detection\data\raw\CelebDF_V3\Celeb-real\Videos_10s"
output_folder = r"C:\Users\miria\Deepfake_Detection\data\processed\CelebDF_V3\Celeb-real\Videos_10s\frames_output"

# --- START ---
# num_videos=2 sorgt dafür, dass nur die ersten 2 genommen werden
extract_frames_from_first_n_videos(input_folder, output_folder, num_videos=50, frame_interval=10)
```
